**Data Cleaning and Pre-processing**

```python
# Data cleaning and pre-processing**

# Handle Missing Values
# Check Missing Values
# Count missing values
df.isnull().sum()
```

```python
# Missing Value Percentage
missing_percent = (df.isnull().sum()/len(df))*100

missing_df = pd.DataFrame({
    'Missing Count': df.isnull().sum(),
    'Missing Percentage': missing_percent
})

missing_df.sort_values('Missing Percentage', ascending=False)
```

```python
# Missing values
df.isnull().sum()

# Duplicate check
df.duplicated().sum()

# Dataset information
df.info()
```

```python
# Drop Missing Description
df.dropna(subset=['Description'], inplace=True)

# Fill Missing CustomerID
df['CustomerID'] = df['CustomerID'].fillna(0)
```

- CustomerID contains the highest number of missing values.

- Missing CustomerID usually represents guest customers.

- Description is important for product-level analysis, so rows with missing descriptions are removed.

- Missing values can reduce analysis accuracy if not handled properly.

```python
# Handle Duplicate Records
# Check Duplicates
duplicate_count = df.duplicated().sum()

print("Duplicate Rows:", duplicate_count)
```

```python
# Remove Duplicates
df.drop_duplicates(inplace=True)

print("Shape after removing duplicates:", df.shape)
```

- Duplicate transactions inflate revenue and sales counts.

- Removing duplicates ensures accurate business reporting.

- Duplicate rows often occur during data entry or system integration.

```python
# Rename Column Names
# Check Columns
df.columns
```

```python
# Rename Columns

df.rename(columns={
    'InvoiceNo':'Invoice_Number',
    'StockCode':'Product_Code',
    'Description':'Product_Description',
    'Quantity':'Quantity_Sold',
    'InvoiceDate':'Invoice_Date',
    'UnitPrice':'Unit_Price',
    'CustomerID':'Customer_ID'
}, inplace=True)
```

- Renaming improves readability.

- Easier coding and reporting.

- Business users understand columns better.

```python
# Drop Unwanted Columns
# Drop if not required
# df.drop(['Product_Code'], axis=1, inplace=True)
```

- Remove irrelevant columns.

- Reduces memory usage.

- Improves model performance.

```python
# Treat Outliers
# Visualize Outliers

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.boxplot(x=df['Quantity_Sold'])
plt.show()
```

```python
# Detect Outliers Using IQR
Q1 = df['Quantity_Sold'].quantile(0.25)
Q3 = df['Quantity_Sold'].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print(lower_limit, upper_limit)
```

```python
# Remove Outliers
df = df[(df['Quantity_Sold'] >= lower_limit) &
        (df['Quantity_Sold'] <= upper_limit)]
df = df[(df['Quantity_Sold'] >= lower_limit) &
        (df['Quantity_Sold'] <= upper_limit)]
df = df[(df['Quantity_Sold'] >= lower_limit) &
        (df['Quantity_Sold'] <= upper_limit)]
```

- Extreme quantities may be bulk purchases or entry errors.

-  Outliers can distort averages and trends.

- IQR method is robust for retail datasets.

```python
# Check Skewness
numeric_cols = ['Quantity_Sold','Unit_Price']

df[numeric_cols].skew()
```

- Retail datasets generally show positive skewness.

- Few large purchases create long right tails.

```python
# Apply Log Transformation
df['Quantity_Sold'].skew()
```

```python

df['Log_Quantity'] = np.log1p(df['Quantity_Sold'])

df['Log_UnitPrice'] = np.log1p(df['Unit_Price'])
```

```python
df[['Log_Quantity','Log_UnitPrice']].skew()
```

- Reduces skewness.

- Makes distribution closer to normal.

- Improves machine learning performance.

```python
# Convert Data Types
df.dtypes
```

```python
df['Customer_ID'] = df['Customer_ID'].astype('int64')
```

```python
df['Invoice_Date'] = pd.to_datetime(df['Invoice_Date'])
```

- Correct datatypes improve processing speed.

- Enables accurate calculations.

- Supports time-series analysis.

Through data cleaning, exploratory data analysis (EDA), KPI tracking, and interactive visualizations, the dashboard transforms raw transactional data into meaningful business insights.

- The analysis revealed that a small group of products and customers contribute significantly to overall revenue, while sales performance varies across countries and time periods. KPI cards, trend analysis, and product-level insights help stakeholders quickly monitor business health and identify growth opportunities.
